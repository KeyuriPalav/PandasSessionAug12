## Series
A Series is a one-dimensional labeled array-like data structure. Series can hold any data type and have an associated index. It is useful for performing operations on a single column or extracting a single dimension from a DataFrame.

* Series is very similar to Numpy array. It is build on top of Numpy array object.
* What differentiate Pandas series from Numpy array is that series can be accessed by index label(has acess labels)


### Creating Series

Creating Series from a list, numpy array, or dictionary.

```python
#list for labels
labels = ['a','b','c'] 

#list of data
my_data = [1,2,3] 
```

```python
# creating pd.Series passing in list of data
pd.Series(my_data)
#default indexing begins with 0; 0,1,2,3,etc
```

```python
# creating pd.Series passing in list of data and labels for indices
s1 = pd.Series(data = my_data, index = labels) #Labelled index Series
s1
```

```python
# another way
pd.Series(my_data,labels)
```

```python
#Creating a Series using Numpy array
arr = np.array(my_data)

type(arr)

pd.Series(arr)
```

## Dataframes

A DataFrame is a two-dimensional labeled data structure that resembles a table or a spreadsheet. It consists of rows and columns, where each column can hold different data types (e.g., numeric, string, boolean). The rows are labeled with an index, and the columns are labeled with column names. DataFrame allows efficient data manipulation, slicing, filtering, merging, and aggregation. It is a powerful tool for data cleaning, exploration, and analysis.

* Dataframes is a collection of series that share the same index

* It consists of rows and columns, where each column can hold different data types (e.g., numeric, string, boolean).
* The rows are labeled with an index, and the columns are labeled with column names.

### Pivot tables
A powerful tool for summarizing data, allowing you to quickly derive insights from your dataset by aggregating and reorganizing it. 

#### Basic Structure of a Pivot Table
The pivot_table() function in pandas takes several parameters to create a pivot table:

* data: The DataFrame to pivot.
* index: The columns you want to group by (rows of the pivot table).
* columns: The columns you want to spread across the top (columns of the pivot table).
* values: The data to aggregate.
* aggfunc: The aggregation function (e.g., mean, sum, count).

```python
pivot = pd.pivot_table(
    merged_df,
    values='PatientID',  # The column to aggregate (count the number of patients)
    index='Ethnicity',   # Group by Ethnicity
    columns='Outcome',   # Separate counts by Outcome (0 or 1)
    aggfunc='count',     # Counting the occurrences
    fill_value=0         # Fill NaN with 0
)

print(pivot)
```
